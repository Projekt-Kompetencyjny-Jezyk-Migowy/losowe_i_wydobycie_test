# 1. Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2. Load data

```python
data = pd.read_excel('złączone_dane.xlsx')
data = data.drop('image_id',axis=1)
data = data.drop(columns=[col for col in data.columns if any (x in col for x in ['3_p', '4_p', '5_p'])])
```

# 3. Data description

```python
pd.set_option('display.max_columns', None)
data.head()
```

# 4. Visualization of data distribution

```python
sns.countplot(x='label', data = data)
plt.title('Data distribution')
plt.show()
```

# 5. Correlation

```python
features_only = data.drop('label',axis=1)
sns.heatmap(features_only.corr(), cmap='coolwarm')
plt.title('Correlation')
plt.show()
```
